# Analyze Experimental Data

Let us load the dataset and take a look at some basic statistics.

```python
from simpoly.poly_arena import experiment

df = experiment.load_data()
```

## Density Analysis

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

ax = axes[0]
ax.hist(
    df["density"],
    bins=20,
    edgecolor="black",
    color="C0",
    alpha=0.7,
    histtype="stepfilled",
)

ax.set_xlabel("Density (g/cm$^3$)")
ax.set_ylabel("Count")

ax = axes[1]
ax.hist(
    df["tg"].dropna(),
    edgecolor="black",
    color="C1",
    alpha=0.7,
    histtype="stepfilled",
)
ax.set_xlabel(r"T$_g$ (K)")

plt.show()
```
